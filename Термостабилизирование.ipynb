```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.special import j0

# Параметры модели
q = 5 # Мощность нагревательнго элемента
alpha = 25.203377569430945 #Коэффициент поверхностной теплоотдачи
cp =  2495040 #Удельная объемная теплоёмкость. Металла???
delta = 3e-3 #Толщина основания узкополосного волоконного лазера
lambda_ = 230  # Коэффициент теплопроводности металла
a = 8.418e-5 #Коэффициент температуропроводности металла
b = 2 * alpha / (cp * delta)  # Коэффициент температуроотдачи

# Радиальные и угловые координаты
max_radius = 100  # Максимальный радиус
num_points = 200  # Количество точек

r = np.linspace(0, max_radius, num_points)
theta = np.linspace(0, 2*np.pi, num_points)
R, Theta = np.meshgrid(r, theta)

# Преобразование в декартову систему для расчета температуры
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Расчет температурного поля
Delta_T = q / (2 * np.pi * lambda_ * delta) * j0((np.sqrt(X**2 + Y**2) * 1e-3) * (b/a)**0.5)

# Построение трехмерного графика
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Поверхностный график
surf = ax.plot_surface(X, Y, Delta_T, cmap='coolwarm')

# Подписи осей
ax.set_xlabel('x, мм')
ax.set_ylabel('y, мм')
ax.set_zlabel('T, K')

# Цветовая шкала
fig.colorbar(surf, shrink=0.5, aspect=10)

# Заголовок
plt.title('Трехмерное распределение температуры по поверхности основания лазера в цилиндрических координатах')

# Сетка
ax.grid(True)

# Показ графика
plt.show()
```

```python
#Рассчёт коэффициента теплоотдачи
print('Если поверхность обдувается, то введите скорость, если не обдувается, то введите 0.0')
w =float(input()) #Скорость воздуха вдоль длины поверхности 
tv = 293 #Температура окружающего воздуха
tp = 295 #Температура поверхности
e = 0.09 #Степень черноты поверхности
B = 0.1 #Ширина поверхности. Нужно уточнить. Степень черноты тела — это энергетическая характеристика излучающего тела, равная отношению потока 
#собственного излучения тела к потоку излучения абсолютно чёрного тела при той же температуре. У аллюминиевого листа e = 0.09
L = 0.1 #Длина поверхности
#Рассчёт проводится для сухого воздуха при атмосферном давлении

t0 = (tv+tp)/2 #Определяющая температура
if w==0.0:
    L0 = (B*L)/(2*(B+L))
else:
    L0 = L
ro = 353.089/(t0+273.15) #Объёмная плотность
a0 = 17.1625
a1 = 4.82102e-2
a2 = -2.17419e-5
a3 = 7.06065e-9
ny = (a0 + a1*t0 + a2*(t0**2) + a3*(t0**3))/1e6 #Динамическая к-т вязкости
b0 = 1.00564e3
b1 = 7.43322e-3
b2 = 5.78429e-4
b3 = -5.87508e-7
b4 = 1.81359e-10
Cp = b0 + b1*t0 + b2*(t0**2) + b3*(t0**3) + b4*(t0**4) #Удельная изобарная теплоёмкость
c0 = 2.41822
c1 = 7.32841e-3
c2 = -2.53698e-6
c3 = 9.34274e-10
Lamba = (c0 + c1*t0 + c2*t0**2 + c3*t0**3)/100 #к-т теплопроводности
v = ny/ro #Кинематическая к-т вязкости
beta = 1/t0 #К-т объёмного расширения
alpha = Lamba/(Cp*ro) #К-т температуропроводности
Pr = ny*Cp/Lamba #Число прандтля
Re = w * L0/v #Число рейнольдса
Gr = 9.8 * beta * abs(tp-tv) * (L0**3) / (v**2) #Число Грасгофа
Ra = Gr*Pr #Число Релея
if w==0.0:
    if (1e4<Ra) and (Ra<1e7):
        Nu = 0.54 * Ra**0.25
    elif (1e7<Ra) and (Ra<1e11):
        Nu = 0.15 * Ra**(1/3)
else:
    if Re<=5*1e5:
        Nu = 0.664 * (Re**0.5) * (Pr**(1/3))
    elif (5*1e5<Re) and (Re<=3*1e7):
        Nu = 0.037 * (Re**0.8) * (Pr**0.43)
#Nu - число Нуссельга 
Alk = Nu*Lamba/L0 #Коэффициент конвективной теплоотдачи
if tv<tp:
    Alp = e*0.00000005670367 * ((tp+273.15)**4 - (tv + 273.15)**4)/(tp-tv)
else:
    Alp = 0.0
#Alp - коэффициент радиационной теплоотдачи
Al = Alk + Alp #Суммарный коэффициент теплоотдачи
q = Al*(tp-tv) #Удельная мощность теплового потока
Q = q*B*L #Полная мощность теплового потока
print(Al)
```
