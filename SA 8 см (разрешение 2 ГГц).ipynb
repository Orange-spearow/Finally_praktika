```python
import math
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
spectr1 = []
Spectr1 = []
with open('8 см.txt', 'r') as file:
    for line in file:
        elements = line.split()
        first_number = float(elements[0])
        second_number = float(elements[1])
        spectr1.append(first_number)
        Spectr1.append(second_number)
file.close()
Spl1 = interp1d(spectr1, Spectr1, kind='cubic')
spectr1_new = np.linspace(spectr1[0], spectr1[-1], 10000)
Spectr1_spl = Spl1(spectr1_new)
plt.figure(figsize=(10,6))
plt.plot(spectr1_new, Spectr1_spl, 'r-')
plt.plot(spectr1, Spectr1, 'ro', label='Исходные данные')
plt.xticks(np.arange(min(spectr1)-0.024-0.0284 , max(spectr1)-0.024-0.0284, 0.2))
plt.xlabel('λ, нм')
plt.ylabel('I, дБм')
plt.title('')
plt.grid(which='major')
plt.show()
H = max(Spectr1_spl)-3
dh = 0.01948712871285352e-9
h = 1549.7718544854486e-9
c = 300000000
G = []
for i in range(len(spectr1_new)):
    if Spectr1_spl[i] - H >= 0:
        G.append(spectr1_new[i])
        #print(spectr1_new[i])
print("Ширина линии на уровне -3 дБм:", G[-1] - G[0], "нм")
#for j in range(len(spectr1_new)):
#    if Spectr1_spl[j] == max(Spectr1_spl):
#        print(spectr1_new[j])
df = c * dh/(1.46 * h**2 )
print("Ширина линии в ГГц",df/1000000000)
```
