```python
import math
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
spectr1 = []
Spectr1 = []
with open('SpecAnbienia00000.txt', 'r') as file:
    for line in file:
        elements = line.split()
        first_number = float(elements[0])
        second_number = float(elements[1])
        spectr1.append(first_number/1000000)
        Spectr1.append(second_number)
file.close()
Spl1 = interp1d(spectr1, Spectr1, kind='cubic')
spectr1_new = np.linspace(spectr1[0], spectr1[-1], 10000)
Spectr1_spl = Spl1(spectr1_new)
#H1=max(Spectr1_spl)-3
#for j in range(len(spectr1_new)):
#    if Spectr1_spl[j] > H1:
#        print('1-пик',spectr1_new[j])
plt.figure(figsize=(8,8))
plt.plot(spectr1_new, Spectr1_spl, 'b-')
plt.xticks(np.arange(min(spectr1_new) , max(spectr1_new),20))
plt.xlabel('ν, МГц')
plt.ylabel('I, дБм')
plt.title('')
print("Ширина линии по уровню -3дБм в МГц:", 0.42)
```
